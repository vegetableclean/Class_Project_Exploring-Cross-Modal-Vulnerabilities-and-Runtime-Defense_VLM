```python
# embeddings_tsne.py
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "openai/clip-vit-large-patch14-336"
processor = CLIPProcessor.from_pretrained(model_id)
model = CLIPModel.from_pretrained(model_id).to(device)

imgs = {
    "target": "../images/meth_pill.png",            # original target you tried to imitate
    "seed": "../images/white.jpeg",                 # starting image
    "adversarial": "../outputs/L2_noNorm_clipgrad_Drug_336_LR0_1-1.jpg"
}

embs = []
labels = []
for name, path in imgs.items():
    im = Image.open(path)
    inp = processor(images=im, return_tensors="pt").to(device)
    emb = model.get_image_features(**inp).detach().cpu().numpy().squeeze()
    embs.append(emb)
    labels.append(name)

embs = np.vstack(embs)  # shape (3, D)
tsne = TSNE(n_components=2, random_state=42, perplexity=1)
proj = tsne.fit_transform(embs)


plt.figure()
for i, label in enumerate(labels):
    x,y = proj[i]
    plt.scatter(x,y)
    plt.text(x+0.5,y+0.5,label)
plt.title("t-SNE of CLIP image embeddings")
plt.show()
```

```python
# perceptual_metrics_fixed.py
"""
Compute perceptual similarity metrics (SSIM and PSNR) between images.
- All images will be resized to the same size (default 336x336)
- Convert to grayscale and normalize to float32 in [0,1]
- When calling skimage.metrics.ssim/psnr, set data_range=1.0
"""

from PIL import Image
import numpy as np
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
import json
import os

def load_gray(path, size=(336, 336)):
    """
    Load an image -> convert to grayscale -> resize -> normalize to [0,1] float32
    """
    if not os.path.exists(path):
        raise FileNotFoundError(f"Image not found: {path}")
    im = Image.open(path).convert('L')        # convert to grayscale
    im = im.resize(size, Image.BILINEAR)     # resize image to match target size
    arr = np.asarray(im).astype(np.float32) / 255.0  # normalize to [0,1]
    return arr

def safe_ssim(a, b):
    """Call SSIM safely, specifying data_range=1.0 for float images"""
    return ssim(a, b, data_range=1.0)

def safe_psnr(a, b):
    """Call PSNR safely, specifying data_range=1.0 for float images"""
    return psnr(a, b, data_range=1.0)

def main():
    # ====== Change to your local paths ======
    target_path = "../images/meth_pill.png"  # target image (the one to mimic)
    adv_path    = "../outputs/L2_noNorm_clipgrad_Drug_336_LR0_1-1.jpg"  # adversarial output
    seed_path   = "../images/white.jpeg"     # initial seed image (white image)
    out_json    = "../outputs/perceptual_metrics_results.json"
    img_size    = (336, 336)  # match CLIP model input size

    # ====== Load images (automatically resized) ======
    try:
        target = load_gray(target_path, size=img_size)
        adv    = load_gray(adv_path, size=img_size)
        seed   = load_gray(seed_path, size=img_size)
    except FileNotFoundError as e:
        print("File error:", e)
        return

    # ====== Compute SSIM and PSNR ======
    results = {}
    try:
        results['ssim_target_vs_adv'] = float(safe_ssim(target, adv))
    except Exception as e:
        results['ssim_target_vs_adv_error'] = str(e)

    try:
        results['psnr_target_vs_adv'] = float(safe_psnr(target, adv))
    except Exception as e:
        results['psnr_target_vs_adv_error'] = str(e)

    try:
        results['ssim_seed_vs_adv'] = float(safe_ssim(seed, adv))
    except Exception as e:
        results['ssim_seed_vs_adv_error'] = str(e)

    try:
        results['psnr_seed_vs_adv'] = float(safe_psnr(seed, adv))
    except Exception as e:
        results['psnr_seed_vs_adv_error'] = str(e)

    # ====== Compute mean absolute difference as extra metric ======
    diff_t_adv = np.abs(target - adv)
    diff_s_adv = np.abs(seed - adv)
    results['mean_absdiff_target_vs_adv'] = float(np.mean(diff_t_adv))
    results['mean_absdiff_seed_vs_adv']   = float(np.mean(diff_s_adv))

    # ====== Print results ======
    print("Perceptual metrics results:")
    for k, v in results.items():
        print(f"  {k}: {v}")

    # ====== Save results to JSON ======
    os.makedirs(os.path.dirname(out_json), exist_ok=True)
    with open(out_json, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2, ensure_ascii=False)

    print(f"\nResult has been saved to: {out_json}")

if __name__ == "__main__":
    main()
```

```python
# embedding_distance.py
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
from torch.nn.functional import normalize

device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "openai/clip-vit-large-patch14-336"
processor = CLIPProcessor.from_pretrained(model_id)
model = CLIPModel.from_pretrained(model_id).to(device)

def emb(path):
    im = Image.open(path)
    inp = processor(images=im, return_tensors="pt").to(device)
    e = model.get_image_features(**inp)
    return normalize(e, dim=-1)

e_target = emb("../images/meth_pill.png")
e_adv = emb("../outputs/L2_noNorm_clipgrad_Drug_336_LR0_1-1.jpg")
e_seed = emb("../images/white.jpeg")

cos_adv_target = (e_adv * e_target).sum().item()
cos_adv_seed = (e_adv * e_seed).sum().item()
print(f"cos(adv, target) = {cos_adv_target:.4f}")
print(f"cos(adv, seed)   = {cos_adv_seed:.4f}")
```

```python
# detect_attack_clip.py
"""
Detect potential adversarial attacks on CLIP model.
- Compare clean and adversarial images
- Compute cosine similarity with a set of text labels
- Flag suspicious similarity drops
- Optionally inspect embedding distances
"""

from PIL import Image
import torch
from transformers import CLIPProcessor, CLIPModel, CLIPTokenizerFast
from sentence_transformers import util
import numpy as np

# ====== CONFIG ======
device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "openai/clip-vit-large-patch14-336"

# Paths to images
clean_img_path = "../images/meth_pill.png"       # Clean/original image
adv_img_path   = "../outputs/L2_noNorm_clipgrad_Drug_336_LR0_1-1.jpg"  # Adversarial image

# Labels to check similarity
labels = ["marijuana", "pill", "drug", "methamphetamine", "LSD"]

# Threshold for flagging attacks
# If similarity to expected label drops below this, we flag it
SIMILARITY_THRESHOLD = 0.7  

# ====== LOAD MODEL ======
processor = CLIPProcessor.from_pretrained(model_id)
tokenizer = CLIPTokenizerFast.from_pretrained(model_id)
model = CLIPModel.from_pretrained(model_id).to(device)

# ====== FUNCTION TO GET IMAGE EMBEDDING ======
def get_img_embedding(img_path):
    img = Image.open(img_path)
    img_tensor = processor(images=img, return_tensors="pt")["pixel_values"].to(device)
    with torch.no_grad():
        emb = model.get_image_features(img_tensor)
    return emb

# ====== FUNCTION TO GET TEXT EMBEDDINGS ======
def get_text_embeddings(labels):
    text_inputs = tokenizer(labels, return_tensors="pt", padding=True).to(device)
    with torch.no_grad():
        text_embs = model.get_text_features(**text_inputs)
    return text_embs

# ====== MAIN ======
def main():
    # Get embeddings
    clean_emb = get_img_embedding(clean_img_path)
    adv_emb   = get_img_embedding(adv_img_path)
    text_embs = get_text_embeddings(labels)

    # Cosine similarities
    clean_sim = util.cos_sim(clean_emb, text_embs)[0].cpu().numpy()
    adv_sim   = util.cos_sim(adv_emb, text_embs)[0].cpu().numpy()

    print("\n=== Similarity with labels ===")
    for i, label in enumerate(labels):
        print(f"{label}: clean={clean_sim[i]:.4f}, adv={adv_sim[i]:.4f}")

    # Check if adversarial similarity drops significantly
    for i, label in enumerate(labels):
        if adv_sim[i] < SIMILARITY_THRESHOLD and clean_sim[i] >= SIMILARITY_THRESHOLD:
            print(f" Potential attack detected on label '{label}'!")

    # Optional: embedding distances
    dist = util.cos_sim(clean_emb, adv_emb).item()
    print(f"\nCosine similarity between clean and adversarial image: {dist:.4f}")

if __name__ == "__main__":
    main()
```
